Beta-lactamase protein, which is the most common in TEM type enterobacteria, is an enzyme that is generally found in gram-negative bacteria and provides resistance to β-lactam antibiotics as a general function (1). Penicillins, cephalosporins, monobactams, and carbapenems are antibiotics in this group, and their common features are the presence of β-lactam rings in their structures and resistance pathways against them. The β-lactam class is one of the most widely used antimicrobial agents in bacterial infection treatment studies with its high effectiveness, low cost and ease of application (2). The β-lactamase enzyme hydrolyzes the β-lactam ring, making these antibiotics inactive. As a result of this situation, it protects the bacteria against antibiotic treatment (1). β-lactamase proteins are grouped into four main classes (A, B, C, D). There is serine amino acid in the active site of classes A, B and C, and zinc ion is found in the active site of class B (3). There are subtypes such as TEM, SHV, CTX-M, OXA. TEM-1 β lactamase is considered the oldest class A wild type and is widely used in many studies (4).

In this study, molecular dynamics simulation of TEM-1 class A β-lactamase protein was performed and analyzed. Based on the analyses, it was discussed how the sequence, structure and dynamics of the protein are related. One of the three-dimensional structures of TEM-1 β-lactamase protein was selected from the Protein Data Bank website (pdb id: 3DTM). This protein, expressed from the _Escherichia coli_ BL21(DE3) organism, is the protein structure of TEM-1 β-lactamase mentioned in the article "Increased folding stability of TEM-1 beta-lactamase by in vitro selection" and kept with the coordinates obtained as a result of that study.

In the published study, the Proside method, an in vitro selection method, was used to find stabilized variants of TEM-1 β-lactamase. Proside, a phage-based method for the selection of thermostable proteins, is used to optimize certain structural and functional properties of the protein and increase its staty (5-70.). Proteins stabilized by proside become resistant to proteases, positively affecting the ability of proside to infect host bacteria. In the study, variants with increased protease resistance of TEM-1 β-lactamase were obtained by increasing the temperature or guanidinium chloride concentration.

### Molecular Dynamics Simulation Analysis

In this section, information about the general function of the protein, its amino acid sequence and comparison with other proteins belonging to the same class as MSA have been provided. At this stage, the output files obtained after the molecular dynamics (MD) simulation of the protein were analyzed. For MD simulations, the steps in the GROMACS tutorial (http://www.mdtutorials.com/gmx/) were followed. In summary, the stages are generate topology, define box and solvate, add ionsi energy minimization, Equilibration, production MD and analysis. After these steps were completed, analyzes were carried out using the files with pdb and dcd extensions. Below are the processing and interpretations of the necessary outputs.

```python
# Required libraries have been downloaded
import logging 
import warnings
from prody import * #prody is a library used to analyze protein structures and dynamics
logger = logging.getLogger('prody')
logger.setLevel(logging.WARNING)
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import tqdm
```

```python
# Load the PDB file 
# and store it in a variable called npt_dry
npt_dry = parsePDB('3DTM_md_simulation_npt_dry.pdb')
# Load the trajectory from the same directory using parseDCD
# and store it in a variable called traj
traj = parseDCD('3DTM_mdsims_output_compact_center_dry.dcd',step=10) 
# Set atoms of traj to the alpha carbonds of npt_dry atoms
traj.setAtoms(npt_dry.select('name CA'))
# Superpose trajectory to the first frame
traj.superpose()
```

The structure and trajectory files, which are the output of the MD simulation, are loaded, the alpha carbon atoms in the trajector are adjusted, and the trajector is aligned according to the first frame by superpose. These processes are used to study the structural flexibility and dynamics of proteins.

```python
# Calculate RMSD of trajectory to the first frame
rmsd = calcRMSD(traj)
```

calcRMSD function RMSD(Root Mean Square Deviation) of the path to the first frame was calculated. RMSD is a commonly used measurement in MD simulations. RMSD is calculated using the formula below.
$$ RMSD = \sqrt{\frac{1}{N} \sum_{i=1}^{N} \| \mathbf{r}_i - \mathbf{r}_i^{\text{ref}} \|^2} $$


Structural similarities or differences between two protein conformations can be determined by RMSD measurement. Proteins move and stretch. RMSD calculates the differences in the position of each atom relative to a reference/starting structure. As a result, the displacements of the atoms are observed. 


```python
# Plot the RMSD
plt.plot(rmsd)
plt.xlabel('Frame')
plt.ylabel('RMSD (angstroms)')
plt.show()
```

A plot was created with the RMSD measurement results of the path leading to the first frame. In the graph, a rapid increase from 0 Å to 1.0 Å is observed in the 0-200 frame range. At this stage, there may be an equilibrium phase in which the protein tries to find its local minimum energy conformation. A slow and steady increase in the range of 1.0-1.5 Å is observed in the 200-600 frame range. At this stage, it can be said that the protein is in a stable state, and since it is in a balanced state, it can be interpreted that there is no conformational change at this stage. In the 600-1000 frame range, 1.5 Å increases to 2-2.5 Å and this is seen as a fluctuation.
To examine the RMSD differences between these frame intervals, the simulation of the structure was watched in the VMD (Visual Molecular Dynamics) program. No significant conformational change was observed in the simulation, but in some regions, differences were observed compared to the beginning of the simulation in the frame range between 600-1000, where the RMSD was highest. These changes may not clearly represent conformational changes. For a better analysis of this situation, the simulation time should be extended and the fluctuations in RMSD values ​​will become more understandable when it is simulated to the point where it reaches the equilibrium state after overcoming the conformational changes and examined again.r


```python
# Loop over the trajectory and calculate the RMSD of each frame to the first frame
# Store the RMSD values in a list called rmsd_list
rmsd_list_ij = []
for i in tqdm.tqdm(range(len(traj))):
    rmsd_list_j = list()
    traj.setAtoms(npt_dry.select('all'))
    traj.setCoords(traj[i].getCoords())
    traj.setAtoms(npt_dry.select('name CA'))
    traj.superpose()
    for j in range(len(traj)):
        rmsd_list_j.append(calcRMSD(traj[i], traj[j]))
    
    rmsd_list_ij.append(rmsd_list_j)

# Convert the list to a numpy array
rmsd_array = np.array(rmsd_list_ij)
```

Traj içindeki her bir çerçeve için döngü kullanılarak RMSD değerleri hesaplandı. Bu kod ile MD simülasyon bopyunca tüm çerçevelerin birbirlerine göre RMSD değerleri hesaplandı ve sonuçlar iki boyutlu bir NumPy dizisine dönüştürülür. Bu işlem yapının zaman içindeki değişikliklerini daha ayrıtnılı şekilde analiz etmek için uygulanmıştır.

```python
# Plot the RMSD array as a heatmap
sns.heatmap(rmsd_array,cmap='Blues')
plt.show()
```

The RMSD heatmap shows how similar structures at different times of the simulation are to each other and visualizes protein conformational diversity and structural changes over time. Dark blue parts of the heatmap indicate high RMSD values, light blue and white colors indicate low RMSD values. In this case, structural differences can be mentioned for regions with high RMSD values, and structural similarities can be mentioned for low RMSD values. In the graph, two different structural situations are observed before and after frame 600. This can be interpreted as a conformational change or fluctuation between different stable states. 500,600 in the previous RMSD chart. There was a significant increase in RMSD value after Frame, and this increase continued until Frame 1000. This was also confirmed by the heatmap graph.
  When the two graphs are examined, it can be said that the protein is more stable in the initial frame regions and becomes more variable after 500-600 frames. It is anticipated that more understandable interpretations of this variability can be made as a result of a longer simulation for the protein.


```python
# Compute the RMSF of the trajectory
rmsf = calcRMSF(traj[200:])
```

Another measurement usually calculated for analysis in MD simulations is RMSF (Root Mean Square Fluctuation), which measures how much each residue moves during the MD simulation. For this process, the RMSF value of each residue was measured for the frames after the 200th frame in the trag. With this measurement, the average movement of the residues was visualized by transferring the values ​​to the plot. Below is the formula for RMSF measurement.


$$ RMSF_i = \sqrt{\frac{1}{T} \sum_{t=1}^{T} (\mathbf{r}_i(t) - \langle \mathbf{r}_i \rangle)^2} $$

```python
# Plot the RMSF
plt.plot(rmsf)
plt.xlabel('Residue')
plt.ylabel('RMSF (angstroms)')
plt.show()
```

The RMSF plot shows for each amino acid how much it deviates from its average positions throughout the simulation. With these values, it can be shown with numerical values ​​which areas in the structure are more flexible. Amino acids with low RMSF values ​​in the graph are stable amino acids that move less. These regions can generally be predicted to be in the core or solid secondary structure elements in the protein. Amino acids with larger RMSF values ​​may indicate more flexible, mobile parts, and these regions are generally more likely to be found in binding sites or on the protein surface.

When the graph is examined, a few residues with very low RMSF values ​​are observed between 0-50 residues. However, values ​​of 1.50 Å are also available. We cannot say that there is a significant flexibility for this region, but it cannot be said that it is completely stable. RMSF values ​​between 50-100 residues tend to increase, with RMSF values ​​increasing up to approximately 2.25 Å. With RMSF values ​​at this height, it can be said that this region is quite flexible.

Between 100-150 residues, the values ​​were generally low and stable. Therefore, it can be said that the residues in this range are more stable. 150-200 residues have quite different RMSF values. Some regions have high RMSF values ​​and some regions have lower RMSF values. It can be interpreted that there are different structural elasticities in different residues for this region.

Among the 200-263 residues, there are residues with low RMSF values ​​and residues with higher values. In general, the highest RMSF values ​​are around 1.25. But the lowest RMSF values ​​are around 0.50. Therefore, it can be said that the RMSF values ​​in this region are quite variable and contain both flexible and stable parts.bilir.

```python
# Perform an essential dynamics analysis on the trajectory
eda = EDA('3DTM_md_simulation')
eda.buildCovariance(traj[200:])
eda.calcModes(n_modes=20)
```

EDA (Essential Dynamics Analysis) is an analysis method used to examine the movement modes of proteins in MD simulations. EDA refers to the examination of data obtained from molecular dynamics simulations using principal component analysis (PCA). This analysis reveals the basic modes of motion of the system and helps understand important structural changes. Here, an EDA object was created, then the covariance matrix was created for the data of the trajector after the 200th clear and the first 20 movement modes were calculated. The covariance matrix shows the movements of atoms relative to each other and calculates the correlation of movements.
The variance of each movement mode was then calculated and rounded to two decimal places. As a result of this calculation, below it is shown how much of the movement each mode represents.

```python
for mode in eda:
    print(calcFractVariance(mode).round(2))
```

```python
showProjection(traj, eda[:2], color='red', marker='.')
plt.show()
```

Tajektori projections according to the first two EDA modes were shown in the graph. These two modes chosen for the EDA chart are the two most important modes. EDA separates the movements of the protein into its main components. Each point represents the protein conformation at a specific time point during the simulation period. Examination of the graph shows that the protein is divided into several different clusters in conformational space. These clusters represent different conformational states of the protein.
The modes on the X and Y axis represent the two main components with the largest variance, and these modes show the main directions of movement of the protein. Distinct clusters in the graph may indicate that the protein transitions between different conformational states and remains in these states for a certain period of time. The wide distribution of coordinates can be associated with the flexibility and mobility of the protein. The different conformational states of the protein observed in this graph are compatible with the previous RMSD and heatmap graphs.

```python
# Compute cross correlation including all modes and plot it as a heatmap
corr = calcCrossCorr(eda)
sns.heatmap(corr,cmap='bwr',center=0)
plt.show()
```

This graph shows correlations between amino acid residues in different regions of the protein. The diagonal line in the graph appears to have a positive correlation value because it shows the correlation of the residue with itself.
When the graph is examined, a significant positive correlation is observed between residues 0-50 and 200-250. Large red regions showing positive correlation may indicate that large protein segments move together, and these regions can be called regions that help maintain structural integrity. Significant negative correlations are observed between residues 150-250 and 50-100. Blue regions showing negative correlation represent that the residues point oppositely and different regions show flexibility.

### Normal Mode Analysis with ANM

Normal Mode Analysis is used to identify flexible states that the protein can reach around its equilibrium position and is a technique that has functional importance in the analysis of a structural study (7). For NMA analysis, which helps to verify the results obtained in MD simulation and interpret the results, the observed motion modes as a result of EDA analysis can be compared with the theoretically predicted motion modes. In this study, Anisotropic Network Model (ANM) was used for Normal mode analysis and the flexibility and dynamic properties of the protein were tried to be understood.

```python
anm = ANM()
system = parsePDB("3DTM.pdb")
anm.buildHessian(system)
anm.calcModes()
writeNMD("3DTM_anm.nmd", anm, system)
```

In the normal mode analysis performed with ANM, the pdb file of the protein obtained at the beginning of the MD simulation was used. The Hessian matrix was created from this file and the modes were calculated and then printed to a .nmd file.
